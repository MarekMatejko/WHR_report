```python
# Importing the libraries we will use
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Reading our csv file with WHR report 
happy_df = pd.read_csv("WHR2023.csv")

# Showing our data 
happy_df
```

```python
# Data types, in our report
happy_df.dtypes
```

```python
# Describing our data
happy_df.describe()
```

```python
# Removing columns, which we will not use
happy_df = happy_df.drop(happy_df.iloc[:,2:5], axis = 1)

happy_df = happy_df.drop(happy_df.loc[:,'Ladder score in Dystopia':'Dystopia + residual'].columns, axis = 1)
```

```python
# Showing data after removing columns
happy_df
```

```python
# Checking if we have some null value
happy_df.isnull().sum()
```

```python
# Finding where we have null value  
happy_df[happy_df['Healthy life expectancy'].isnull()]
```

```python
# Showing correlation between columns
corr = happy_df.corr()
plt.subplots(figsize=(12,8))
heat = sns.heatmap(corr,vmin = -1,cmap="coolwarm", fmt='.0%', annot=True)
```

```python
# Finding mean and median for ladder score
mean_happy = np.array(happy_df['Ladder score']).mean()
median_happy = np.median(happy_df['Ladder score'])
```

```python
# Showing how ladder score is changing in table
plt.figure(figsize=(14,8))
plt.title("Ladder score from worst to best", fontsize = 20)
plt.xlabel('Ladder score')
plt.ylabel('Index')
plt.axvline(mean_happy, color = 'red',label = "Mean %s" % (round(mean_happy,2)))
plt.axvline(median_happy, color = 'green',label = "Median %s" %(round(median_happy,2)) )
plt.legend(loc="lower left",fontsize=14)
plt.plot(np.array(happy_df['Ladder score']), happy_df.index)
```
