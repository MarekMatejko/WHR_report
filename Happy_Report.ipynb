```python
# Importing the libraries we will use
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# Reading our csv file with WHR report 
happy_df = pd.read_csv("WHR2023.csv")

# Showing our data 
happy_df
```

```python
# Data types, in our report
happy_df.dtypes
```

```python
# Describing our data
happy_df.describe()
```

```python
# Removing columns, which we will not use
happy_df = happy_df.drop(happy_df.iloc[:,2:5], axis = 1)
happy_df = happy_df.drop(happy_df.loc[:,'Ladder score in Dystopia':'Dystopia + residual'].columns, axis = 1)
```

```python
# Renaming column 'Ladder score' to 'Happiness score'
happy_df = happy_df.rename({'Ladder score':'Happiness score'}, axis = 1)
happy_df
```

```python
# Showing data after removing columns
happy_df
```

```python
# Checking if we have some null value
happy_df.isnull().sum()
```

```python
# Finding where we have null value  
happy_df[happy_df['Healthy life expectancy'].isnull()]
```

```python
# Checking if we have any duplicated
happy_df["Country name"].duplicated().sum()
```

```python
# Showing correlation between columns
corr = happy_df.corr()
plt.subplots(figsize=(12,8))
heat = sns.heatmap(corr,vmin = -1,cmap="coolwarm", fmt='.0%', annot=True)
```

```python
# Finding mean and median for happiness score
mean_happy = np.array(happy_df['Happiness score']).mean()
median_happy = np.median(happy_df['Happiness score'])
min_happy = np.amin(happy_df['Happiness score'])
max_happy = np.amax(happy_df['Happiness score'])
```

```python
# Showing how ladder score is changing in table
plt.figure(figsize=(14,8))
plt.title("Happiness score from worst to best", fontsize = 20)
plt.xlabel('Happiness score')
plt.ylabel('Index')
plt.axvline(min_happy, color = 'black',label = "Minmal score %s" %(round(min_happy,2)))
plt.axvline(max_happy, color = 'green',label = "Maximal score %s" %(round(max_happy,2)))
plt.legend(loc="lower right",fontsize=14)
plt.plot(np.flip(happy_df['Happiness score']), happy_df.index)
```

```python
# Creating histogram with mean and median.
fig, ax = plt.subplots(figsize=(12,6))
hist = ax.hist(happy_df['Happiness score'],bins = range(0,10,1), ec = "white")
ax.axvline(mean_happy, color = 'red',label = "Mean %s" % (round(mean_happy,2)))
ax.axvline(median_happy, color = 'orange',label = "Median %s" %(round(median_happy,2)))
ax.legend(loc="upper right",fontsize=12)
plt.xticks(range(0,10,1))
plt.title("Distribution happiness score for countries", fontsize = 20)
plt.xlabel('Happiness score')
plt.ylabel('Countries')

# Showing number of countries in bins
labels = [hist[0][i] for i in range(len(hist[0]))]
rects = ax.patches
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(
        rect.get_x() + rect.get_width() / 2,
        height + 0.1,
        int(label),
        ha='center',
        va='bottom'
    )
```

```python
# Looking for countries that were in both reports
same = set(happy_df['Country name']) & set(happy_df_2022['Country'])
list(same)
```

```python
# Creating data with this same countries in both reports
happy_score_2023 = pd.DataFrame()
happy_score_2022 = pd.DataFrame()
for i in same:
    happy_score_2022 = pd.concat([happy_score_2022, happy_df_2022[(happy_df_2022['Country'] == i)]])
    happy_score_2023 = pd.concat([happy_score_2023, happy_df[(happy_df['Country name'] == i)]])
    
# Leaving only columns with countries and happiness score
happy_score_2022 = happy_score_2022.drop(happy_score_2022.iloc[:,0:1], axis = 1)
happy_score_2023 = happy_score_2023.drop(happy_score_2023.iloc[:,2:], axis = 1)    
    
# Sorting from the best happiness score
happy_score_2023 = happy_score_2023.sort_values(by=['Happiness score'], ascending=False, ignore_index=True)
happy_score_2022 = happy_score_2022.sort_values(by=['Happiness score'], ascending=False, ignore_index=True)


```
